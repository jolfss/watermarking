```python
# Methods to append ANSI-Terminal escape codes for coloring.
def colorify(s:str, fg:Optional[Tuple[int,int,int]],bg:Optional[Tuple[int,int,int]]) -> str:
  """Makes text colored where fg,bg are foreground and background color given as a (r,g,b) [0-255]."""
  sfg = f"38;2;{fg[0]};{fg[1]};{fg[2]}" if fg else ""
  sbg = f"48;2;{bg[0]};{bg[1]};{bg[2]}" if bg else ""
  if colorcode := (sfg + sbg) != "":
    return f"\033[0;{sfg};{sbg}m" + s + "\033[0m"
  return s

# Color shortcuts
red = lambda s : colorify(s,(0,0,0),(245,185,185))
green = lambda s : colorify(s,(0,0,0),(175,245,185))
bold = lambda s : f"\033[1m{s}\033[0m"

 # NOTE: The below method was created by GPT-4
def printnl(text, max_length=80):
  """Prints the text with newlines inserted to ensure that no line exceeds the specified maximum length.
  :param text: The text to be printed.
  :param max_length: The maximum length of each line."""
  words = text.split()  # Split the text into words
  current_line = ""  # Initialize the current line as empty

  for word in words:
      # Check if adding the next word would exceed the max length
      if len(current_line) + len(word) + 1 > max_length:
          print(current_line)  # Print the current line
          current_line = word  # Start a new line with the current word
      else:
          if current_line:  # If the line is not empty, add a space before the word
              current_line += " "
          current_line += word  # Add the word to the current line

  # Print any remaining text in the last line
  if current_line:
      print(current_line)
```

### Implementation of Watermark

```python
# NOTE: It looks like three main classes comprise the base implementation.
# https://github.com/jwkirchenbauer/lm-watermarking/blob/main/watermark_reliability_release/watermark_processor.py
"""
Cited external resources:
[1] https://huggingface.co/docs/transformers/v4.39.3/en/internal/generation_utils#transformers.LogitsProcessor
[2] https://huggingface.co/docs/transformers/v4.39.3/en/main_classes/pipelines#transformers.Text2TextGenerationPipeline
[3] https://huggingface.co/tasks/text-generation
"""

#------------------------#
#   Watermark Settings   #
#------------------------#
class WatermarkBase:
  """This class serves as an interface for the LLM-Watermarking protocol described in the paper."""
  def __init__(self,vocab:list[int], # V, list of tokens
               gamma:float=0.5,      # γ ∈ (0,1)
               delta:float=2         # δ > 0
               ):
    # V, vocabulary
    self.vocab = vocab
    self.vocab_size = len(vocab)
    """..."""

    # γ, gamma
    self.gamma = gamma
    self.greenlist_size = int(self.gamma*self.vocab_size)
    """This parameter determines the green-list size, it is the proportion of
    green-listed tokens. There are γ|V| green-words and (1-γ)|V| red-words."""

    # δ, delta
    self.delta = delta
    """This is the 'hardness' parameter---it is the constant which gets added to
    the raw logits of green tokens to boost their signal."""

    # rng, set up red/green words
    self.greenlist_ids = torch.randperm(self.vocab_size,dtype=torch.long)[:self.greenlist_size]
    mask = torch.FloatTensor(torch.zeros(len(vocab), dtype=torch.float))
    mask[self.greenlist_ids] = 1.0
    self.greenlist_mask = mask
    """NOTE: The original paper includes some cryptographic measures and parametrizes the seed on
    some amount of input sequence; we are not worried about such concerns here so we opt to use a
    constant greenlist."""


#------------------------------------------------------------#
#   Watermarker (impl. of HF LogitsProcessor for pipeline)   #
#------------------------------------------------------------#
class WatermarkLogitsProcessor(WatermarkBase, LogitsProcessor):
  """This is an implementation of the WatermarkBase class which supports the LogitsProcessor interface,
  which is a HF abstraction which can be used in generation pipelines (transformers.Pipeline)
  More details can be found in [1]."""
  def __init__(self, *args, **kwargs):
    super().__init__(*args, **kwargs)
    self.counter=0

  def __call__(self,
               input_ids: torch.LongTensor,# (batch_size, sequence_length)
               #^^^^^^^^^ NOTE: The original authors use the input to parametrize the green words,
               # we simply have these generated as constants upon creation.
               scores: torch.Tensor # [dtype=float] (batch_size, config.vocab_size) NOTE: config refers to the pipeline
               ) -> torch.Tensor:   # [dtype=float] (batch_size, config.vocab_size)
    """Applies the watermark bias to the logits for the next token.
    Parameters:
      input_ids: Indices of input sequence tokens in the vocabulary.
      scores: Prediction scores of a language modeling head. These can be logits for each vocabulary
        when not using beam search or log softmax for each vocabulary token when using beam search.
    Returns:
      The processed prediction scores.
    -----------------documentation taken from [1]-----------------"""
    expanded_mask = self.greenlist_mask.to(scores.device).expand_as(scores)
    scores += self.delta * expanded_mask

    return scores # NOTE: Requires dtype=float

#---------------------------------------------------------#
#   Watermark Detector (currently requires watermarker)   #
#---------------------------------------------------------#
class WatermarkDetector(WatermarkBase):
  """This class serves as an interface for the LLM-Watermarking protocol described in the paper."""
  def __init__(self, *args,
               watermarker: WatermarkLogitsProcessor,
               tokenizer: Tokenizer,
               z_threshold: float = 4.0,
               # NOTE: We have access to the watermarker so we will just hardwire the analysis.
               # device : torch.device,
               # NOTE: We are not evaluating adversariality so we do not need to implement the
               # normalizers/etc. which they examine.
               **kwargs):
    super().__init__(*args, **kwargs)

    self.device = device
    self.tokenizer = tokenizer
    self.z_threshold = z_threshold

    self.watermarker = watermarker
    self.greenlist_ids = watermarker.greenlist_ids #NOTE: Right now just copies from generator.

  def __call__(self, seq: Union[list[int], str]):
      """Detect the watermark."""
      if isinstance(seq, str):
          tokens = self.tokenizer.encode(seq)
      elif isinstance(seq, list):
          tokens = seq
      else:
          raise ValueError("Input must be a string or a list of token IDs.")

      # Visualize each token's watermark status
      # return self.visualize_watermark(tokens)
      return self.z_score(tokens) > self.z_threshold

  def visualize_watermark(self, seq: Union[list[int], str]):
      """Visualizes tokens to check if they are part of the greenlist."""
      if isinstance(seq, str):
          tokens = self.tokenizer.encode(seq)
      elif isinstance(seq, list):
          tokens = seq
      else:
          raise ValueError("Input must be a string or a list of token IDs.")
      output = []
      for token in tokens:
          token_str = self.tokenizer.decode([token])
          token_str = green(token_str) if token in self.watermarker.greenlist_ids else red(token_str)
          output.append(token_str)
      return ''.join(output)

  def z_score(self, tokens: list[int]):
    """Performs a one proportion z-test to evaluate the null hypothesis,
    which states that the sequence is generated with no knowledge of the
    green list."""
    green_count = 0
    total_count = len(tokens)
    for token in tokens:
      if token in self.greenlist_ids:
        green_count += 1
    expected_count = self.gamma * total_count
    z = (green_count - expected_count)
    z /= numpy.sqrt(expected_count * (1 - self.gamma))
    return z
```

### Set up experiment

```python
from dataclasses import dataclass

@dataclass(frozen=True)
class ExpConfig:
  gamma: float
  delta: float
  z: float

@dataclass(frozen=True)
class ExpResult:
  # True if watermark was detected on marked text
  detected_when_marked: bool
  # True if watermark was detected on unmarked text
  detected_when_not_marked: bool

def setup_experiment(config: ExpConfig) -> Tuple[WatermarkLogitsProcessor, WatermarkDetector]:
  vocab = tokenizer.get_vocab()
  watermarker = WatermarkLogitsProcessor(vocab=vocab, gamma=config.gamma, delta=config.delta)
  detector = WatermarkDetector(vocab=vocab, watermarker=watermarker, tokenizer=tokenizer, z_threshold=config.z)
  return watermarker, detector

def statistics(watermark_detection_results, no_watermark_detection_results):
  """Returns (false positive, true negative, false negative, true positive) rates as a tuple."""
  fp, fn, tp, tn = 0, 0, 0, 0
  for detected in watermark_detection_results:
    if detected:
      tp += 1
    else:
      fn += 1
  for detected in no_watermark_detection_results:
    if detected:
      fp += 1
    else:
      tn += 1
  fpr = fp / (fp + tn)
  tnr = tn / (fp + tn)
  fnr = fn / (fn + tp)
  tpr = tp / (fn + tp)
  return (fpr, tnr, fnr, tpr)

def print_results(results: Dict[ExpConfig, List[ExpResult]], latex=False):
  if not latex:
    print(bold("Watermarked, Not Watermarked"))
    for exp, exp_results in results.items():
      watermarked_tally = []
      not_watermarked_tally = []
      for expresult in exp_results:
        wm, nwm = expresult.detected_when_marked, expresult.detected_when_not_marked
        watermarked_tally.append(wm)
        not_watermarked_tally.append(nwm)
      fpr, tnr, fnr, tpr = statistics(watermarked_tally, not_watermarked_tally)
      print(f"Results from {bold(f'experiment={exp}')} fpr:{fpr}, tnr:{tnr}, tpr:{tpr}, fnr:{fnr}")
  else:
    print(r"$$\begin{matrix}")
    print(r"&&&& z=4 &&&& z=5 \\")
    print(r"sampling & \delta & \gamma & count &  FPR &  TNR &  TPR &  FNR &  FPR &  TNR &  TPR &  FNR \\")
    # Sorting by z-score in ascending order, and then by delta, gamma (gamma in descending order)
    sorted_results = sorted(results.items(), key=lambda item: (item[0].z, item[0].delta, item[0].gamma * -1))
    # map of (delta,gamma) to (results with z=4, results with z=5)
    dg_to_z = {(res[0].delta, res[0].gamma): {"z=4": None, "z=5": None} for res in sorted_results}
    for res in sorted_results:
      assert res[0].z == 4 or res[0].z == 5, f"Failed format z-score in table: {res[0].z=}"
      dg_to_z[(res[0].delta, res[0].gamma)]["z=4" if res[0].z == 4 else "z=5"] = res
    for (delta, gamma), z_results in dg_to_z.items():
      (z_4_config, z_4_results), (z_5_config, z_5_results) = z_results["z=4"], z_results["z=5"]
      assert z_4_config.z == 4 and z_5_config.z == 5 and z_4_config.delta == z_5_config.delta and z_4_config.gamma == z_5_config.gamma

      z4_fpr, z4_tnr, z4_fnr, z4_tpr = statistics([res.detected_when_marked for res in z_4_results], [res.detected_when_not_marked for res in z_4_results])
      z5_fpr, z5_tnr, z5_fnr, z5_tpr = statistics([res.detected_when_marked for res in z_5_results], [res.detected_when_not_marked for res in z_5_results])
      print(f"\\text{{m-nom. }} & {delta} & {gamma} & {len(z_4_results)} & {z4_fpr:.3f} & {z4_tnr:.3f} & {z4_tpr:.3f} & {z4_fnr:.3f} & {z5_fpr:.3f} & {z5_tnr:.3f} & {z5_tpr:.3f} & {z5_fnr:.3f}", end='\\\\ \n')

    print(r"\end{matrix}$$")


def run_batched_experiments(prompts: List[str], batch_size=4) -> Dict[ExpConfig, List[ExpResult]]:
  """Returns a mapping of experiment parameters to each prompt's watermarked and unwatermarked results."""
  experiments = [ExpConfig(gamma, delta, z) for z in (4,5) for delta in (1.0, 2.0, 5.0) for gamma in (0.25, 0.5)]
  experiment_results = {}
  for batch in tqdm(range(len(prompts)//batch_size)):
    for experiment_config in tqdm(experiments):
      for (prompt_id, prompt) in enumerate(prompts[batch_size*batch : min(len(prompts),batch_size*(batch+1))]): # NOTE: used to be max...smh
        # get watermarker and its detector
        watermarker, detector = setup_experiment(experiment_config)
        z, delta, gamma = experiment_config.z, experiment_config.delta, experiment_config.gamma

        # tokenize prompt
        input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)

        # generate [un]watermarked texts
        texts = {}
        watermark_processors = {"unwatermarked": [], "watermarked": [watermarker]}
        for setting, processors in watermark_processors.items():
          tokens = model.generate(
              input_ids,
              pad_token_id=tokenizer.eos_token_id,
              min_length=195 + input_ids.size(-1),
              max_length=205 + input_ids.size(-1),
              num_return_sequences=1,
              do_sample=True,
              logits_processor=LogitsProcessorList(processors)
          )
          text = tokenizer.decode(tokens[0][input_ids.size(-1):], skip_special_tokens=True)
          # print(f"Text length={len(tokens[0][input_ids.size(-1):])}")
          texts[setting] = text

        # pass texts through our detector
        watermarked_result = detector(texts["watermarked"])
        not_watermarked_result = detector(texts["unwatermarked"])
        result = ExpResult(detected_when_marked=watermarked_result, detected_when_not_marked=not_watermarked_result)

        # gets result list @experiment_config or initializes it to [] and then appends
        experiment_results.setdefault(experiment_config, []).append(result)

    print_results(experiment_results, latex=True)
  return experiment_results
```

```python
from statistics import mean

def get_mean_length(n_inputs):
  """Returns average generated token length. A utility function we occasionally used to debug"""
  watermarker, detector = setup_experiment(ExpConfig(0.25,1,4))
  all_texts = []
  all_tokens = []
  for prompt in tqdm(input_prompts[:n_inputs]):
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)
    # generate watermarked texts
    tokens = model.generate(
        input_ids,
        pad_token_id=tokenizer.eos_token_id,
        min_length=195 + input_ids.size(-1),
        max_length=205 + input_ids.size(-1),
        num_return_sequences=1,
        do_sample=True,
        logits_processor=LogitsProcessorList([watermarker])
    )
    all_tokens.append(tokens)
    text = tokenizer.decode(tokens[0], skip_special_tokens=True)
    all_texts.append(text)
  return mean([all_tokens[0].size()[1] for t in all_tokens])

get_mean_length(2)
```
